 User Overview analysis 



top 10 handsets used by the customers

```python
query = ''' 
    SELECT DISTINCT 
        "Handset Type", 
        COUNT(*) as UsageCount
    FROM df
    GROUP BY "Handset Type" 
    ORDER BY UsageCount DESC
    limit 10
'''

output = pysqldf(query)
output
```

 top 3 handset manufacturers

```python
query ='    SELECT DISTINCT "Handset Manufacturer", COUNT(*) as "Number of Users"FROM df  GROUP BY "Handset Manufacturer" ORDER BY "Number of Users" DESC limit 3'
output = pysqldf(query)
output
```

 identify the top 5 handsets per top 3 handset manufacturer

```python
query ='''SELECT `Handset Manufacturer`, `Handset Type`
FROM (
    SELECT `Handset Manufacturer`, `Handset Type`,
           ROW_NUMBER() OVER (PARTITION BY `Handset Manufacturer` ORDER BY COUNT(*) DESC) AS rn
    FROM df
    GROUP BY `Handset Manufacturer`, `Handset Type`
) AS subquery
WHERE rn <= 5
ORDER BY `Handset Manufacturer`, rn'''

output = pysqldf(query)
output
```

number of xDR sessions


```python
query = '''SELECT `MSISDN/Number` AS User, COUNT(*) AS SessionCount
FROM df
GROUP BY `MSISDN/Number`'''
output = pysqldf(query)
output
```

total download and upload

```python
query ='''SELECT `MSISDN/Number` AS User, SUM(`Total DL (Bytes)`) AS TotalDownload, SUM(`Total UL (Bytes)`) AS TotalUpload
FROM df
GROUP BY `MSISDN/Number`'''
output = pysqldf(query)
output
```

Session Duration

```python
query ='''SELECT `MSISDN/Number` AS User, SUM(`Dur. (ms)`) AS SessionDuration
FROM df
GROUP BY `MSISDN/Number`'''

output = pysqldf(query)
output
```
